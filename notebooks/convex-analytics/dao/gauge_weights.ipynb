```python
import requests
```

```python
votes_query = """query Votes {
  votes (
    first: 1000
    skip: 0
    where: {
      proposal: "QmaS9vd1vJKQNBYX4KWQ3nppsTT3QSL3nkz5ZYSwEJk6hZ"
    }
    orderBy: "created",
    orderDirection: desc
  ) {
    id
    voter
    created
    proposal {
      id
    }
    choice
    space {
      id
    }
  }
}
"""

proposals_query = """query Proposal {
  proposal(id:"QmaS9vd1vJKQNBYX4KWQ3nppsTT3QSL3nkz5ZYSwEJk6hZ") {
    id
    title
    body
    choices
    start
    end
    snapshot
    state
    author
    space {
      id
      name
    }
  }
}
"""
```

```python
url = "https://hub.snapshot.org/graphql"
```

# Proposal

```python
r = requests.post(url, json={'query': proposals_query})
proposal_data = dict(r.json())
proposal_data['data']
```

# Votes Analysis

```python
r = requests.post(url, json={'query': votes_query})
votes_data = dict(r.json())
votes_data['data']
```
