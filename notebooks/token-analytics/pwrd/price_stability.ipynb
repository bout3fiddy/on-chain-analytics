```python
import pandas as pd
from pycoingecko import CoinGeckoAPI

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
```

```python
coingecko = CoinGeckoAPI()
```

```python
pwrd_price = pd.DataFrame(data=coingecko.get_coin_market_chart_by_id(id="pwrd-stablecoin", vs_currency="usd", days="max")['prices'], columns=["time", "price_usd"])
time_start = pwrd_price.time[0]/1000
time_end = pwrd_price.time.iloc[-1]/1000
pwrd_price['time'] = pd.to_datetime(pwrd_price['time']/1000, unit='s')
```

```python
fig, ax = plt.subplots(1,1, figsize=(25, 8))

ax.plot(
    pwrd_price['time'],
    pwrd_price['price_usd'],
    color='red',
    label='PWRD',
    lw=1
)

ax.set_ylabel("Price [$]", fontsize=25)
ax.set_xlabel("", fontsize=25)
ax.tick_params(axis='both', which='major', labelsize=25, colors='black')
ax.yaxis.get_major_formatter().set_scientific(False)
ax.yaxis.get_major_formatter().set_useOffset(False)
ax.legend(fontsize=25);
```
