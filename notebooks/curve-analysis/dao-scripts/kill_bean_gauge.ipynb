```python
import os, sys
sys.path.append('../../../')

import warnings

import brownie

from utils import simulate_curve_dao_vote, CURVE_DAO_OWNERSHIP
from utils import configure_network
from utils import init_contract

warnings.filterwarnings("ignore")
```

Connect to mainnet-fork ...

```python
ALCHEMY_API_KEY = os.environ['ALCHEMY_API_KEY']
configure_network(
    node_provider_https=f"https://eth-mainnet.alchemyapi.io/v2/{ALCHEMY_API_KEY}",
    network_name='mainnet'
)
brownie.network.connect('mainnet-fork')
```

Set up governance vote inputs. For description of individual vars, check curve_dao_utils.py in the utils folder.

```python
TARGET = CURVE_DAO_OWNERSHIP
GAUGE_TO_KILL = "0x6F98dA2D5098604239C07875C6B7Fd583BC520b9"
FACTORY_ADMIN = "0x2EF1Bc1961d3209E5743C91cd3fBfa0d08656bC3"
ACTIONS = [
    (FACTORY_ADMIN, "set_killed", GAUGE_TO_KILL, True),
]
DESCRIPTION = "Kill BEAN:3CRV Gauge."
```

Simulate Gauge Killing

```python
simulate_curve_dao_vote(target=TARGET, actions=ACTIONS, description=DESCRIPTION)
```

Test if gauge was killed successfully

```python
gauge_contract = init_contract(GAUGE_TO_KILL)
assert gauge_contract.is_killed()
```
